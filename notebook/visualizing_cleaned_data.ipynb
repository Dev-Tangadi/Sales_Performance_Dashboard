```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the cleaned data set
df = pd.read_csv(r"C:\Users\Dev\Desktop\analyst project\output\cleaned_superstore.csv",encoding="ISO-8859-1")
```

```python
df.info()
```

```python
sales_category = df.groupby("Category")["Sales"].sum().sort_values(ascending =False)
```

```python
#visualizing the data
plt.figure(figsize=(8,5))
sns.barplot(x= sales_category.index,y= sales_category.values,palette='viridis')
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()
```

```python
profit_subcat = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending =False)
```

```python
#visualizing the data
plt.figure(figsize=(10,5))
sns.barplot(x= profit_subcat.index,y= profit_subcat.values,palette='coolwarm')
plt.title("Total Profit By Sub Category")
plt.ylabel("Profit")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Monthly sales trends

monthly_sales = df.groupby("Month")["Sales"].sum()
#visualizing the data
plt.figure(figsize=(8,5))
sns.barplot(x=monthly_sales.index,y= monthly_sales.values,palette='viridis')
plt.title("Monthly sales trend")
plt.ylabel("Sales")
plt.xlabel("Months")
plt.show()
```

```python
#Region wise sales
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

print(region_sales)

```

```python
#visualize the region sales

plt.figure(figsize = (8,5))
sns.barplot(x= region_sales.index,y =region_sales.values,palette="magma")
plt.title("Region Sales")
plt.ylabel('Sales')
plt.show()
```

```python
#Monthly Region wise sales
monthly_region_sales = df.groupby(["Month","Region"])["Sales"].sum().sort_values(ascending=False)

print(monthly_region_sales)
```

```python
monthly_region_sales.plot(figsize=(8,5),marker='o')
plt.title("Monthly Sales Region wise")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

```python
#Top 10 profitable Products

top_products = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)

print(top_products)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=top_products.index,y=top_products.values,palette="summer")
plt.title("Top 10 Profitable Products")
plt.ylabel("profit")
plt.xticks(rotation = 90)
plt.show()
```
